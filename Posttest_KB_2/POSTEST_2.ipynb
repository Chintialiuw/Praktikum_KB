## Mengimport Library Pandas untuk Membaca File CSV

```python
import pandas as pd
Books = pd.read_csv("top2k_book_descriptions.csv")
```

## Menampilkan jumlah record, nama attribute, jumlah attribute dengan tipe data objek dan angka serta Menampilkan jumlah attribute yang memiliki nilai null

```python
Books.info()
```

## Memasukkan informasi ke dalam dictionary dengan key sesuai nama informasinya

```python
record, attribute = Books.shape
jumlah_angka = Books.select_dtypes(["int64", "float64"]).any().sum()
jumlah_object = Books.select_dtypes(["object"]).any().sum()
nama_attribute = Books.columns.values
jumlah_null = Books.isnull().any().sum()

hasil = {
    "Jumlah Record Keseluruhan" : record,
    "\nJumlah Attribute Keseluruhan" : attribute,
    "\nNama Attribute" : list(nama_attribute),
    "\nJumlah Attribute (Tipe Angka)" : jumlah_angka,
    "\nJumlah Attribute (Tipe Object)" : jumlah_object,
    "\nJumlah Nilai Null" : jumlah_null
}

for x in hasil:
    print(x, ":", hasil[x])
```

## Menampilkan rata-rata, standar deviasi, nilai minimun, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka

```python
Books.describe()
```

## Menampilkan nilai korelasi attribute bertipe angka

```python
Books.corr()
```

## Mengubah niai salah satu attribute bertipe angka menjadi desimal 3 angka di belakang koma

```python
Books["ratings_5"] = Books["ratings_5"].astype("float64")
Books["ratings_5"] = Books["ratings_5"].apply("{:.3f}".format)
Books[["ratings_5"]]
```
