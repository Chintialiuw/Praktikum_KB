Mengimport Library Pandas, Matplotlib, dan Seaborn

```python
import pandas as pd
from matplotlib import pyplot as plt

import seaborn as sns
```

Membaca File CSV dan Info di dalam Dataset

```python
df = pd.read_csv("netflix daily top 10.csv")
df.info()
```

Menampilkan Isi dari File Dataset

```python
df.head()
```

Membuat Grafik Batang tentang Viewership Score dan Type

"Jenis Acara yang paling banyak dilihat atau ditonton oleh banyak orang adalah TV Show"

```python
x = list(df["Type"])
y = list(df["Viewership Score"])

plt.figure(figsize = (10,5))
plt.bar(x, y, label = "Type", color = "red")
plt.title("Peminat Jenis Acara")
plt.legend()
plt.show()
```

Membuat Grafik Boxplot tentang Type dan Days In Top 10

"Setiap harinya Jenis Acara seperti TV Show sangat digemari oleh para penonton dan pada grafik ini menunjukkan bahwa data miningnya tidak simetris dikarenakan median di dalam kotak tidak berada di tengah dan garis atas lebih panjang dibandingkan dengan garis bawah, sehingga terjadi kecenderungan miring ke kanan (positive shewness)"

```python
sns.boxplot(data = df, x = 'Type', y = 'Days In Top 10')
```

Membuat Grafik Pie tentang Type

Mencari berapa nilai yang muncul di dalam setiap index "type" berada

```python
data_pie = df['Type'].value_counts().rename('Type').reset_index(name = 'id_count')
data_pie
```

"Grafik di bawah ini menunjukkan bahwasannya TV Show sangatlah banyak diminati yakni sekitar 62.62% dibandingkan Jenis Acara lainnya, berbanding terbalik dengan Jeni Acara Concert/Performance yang memiliki sedikit peminat yakni sekitar 0.03%"

```python
plt.pie(data_pie.id_count, 
        labels = data_pie['index'], 
        startangle = 170, 
        explode = (0.01,0.02,0.2,0.15),
        autopct ='%.2f%%', 
        shadow = False)

plt.title('Kecenderungan Penonton terhadap Jenis Acara')
plt.show()
```
